```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score
import unicodedata
```

```python
# Read in labels

with open('/content/drive/MyDrive/CSE 573: Fake Reviews Project/YelpZip/metadata') as file:
  label_data = file.readlines()

# Remove last character of newline
label_data = [row[:-1] for row in label_data]

Y = [row[:-1].split('\t')[3] for row in label_data]
```

```python
transformer = TfidfVectorizer(stop_words='english', max_features=10000)
dataset = transformer.fit_transform(data)
```

```python
# Check size of dataset
dataset.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(dataset, Y, random_state=42,train_size=0.8) #train 80% test 20%

```

```python
# Train and test classifier
rdf_classifier = RandomForestClassifier(n_estimators=100, n_jobs=-1)
rdf_classifier.fit(x_train, y_train)

```

```python
y_pred = rdf_classifier.predict(x_test)
```
